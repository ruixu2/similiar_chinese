```python
import cv2  
import numpy as np  
import os
from skimage.metrics import structural_similarity as ssim
import skimage as ski
```

```python
def get_similiar(path1,path2):  
    image1 = ski.io.imread(path1)
    image2 = ski.io.imread(path2)

    print(image1.shape,image2.shape)
    # 计算结构相似性指数
    similarity_index = ssim(image1, image2)

    return similarity_index
```

```python
chinese_list=[]
for i in range(0x4E00, 0x9FA5 + 1):
    temp=chr(i)
    chinese_list.append((i,temp))
chinese_list
```

```python
for item in chinese_list:
    idx=item[0]
    i=item[1]
    img_path1=f"../img/{idx}"+"_"+i+".jpg"
    for item2 in chinese_list[idx+1:]:
        idx2=item2[0]
        i2=item2[1]
        img_path2=f"../img/{idx2}"+"_"+i2+".jpg"
        print(img_path1,img_path2)
        similarity=get_similiar(img_path1,img_path2)
        print(similarity)
        break
```
